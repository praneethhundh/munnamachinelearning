```python
#ex-1
import pandas as pd

url="https://raw.githubusercontent.com/kevinadhiguna/find-S-algorithm/refs/heads/master/data.csv"
data = pd.read_csv(url)

examples = data.values.tolist()

def find_s_algorithm(examples):
    hypothesis = examples[0][:-1]  

    for example in examples:
        if example[-1] == "Yes": 
            for i in range(len(hypothesis)):
                if hypothesis[i] != example[i]:
                    hypothesis[i] = "?"  
    return hypothesis

final_hypothesis = find_s_algorithm(examples)
print("Final Hypothesis:", final_hypothesis)
```

```python
#ex-2
import csv

with open("trainingexamples.csv") as f:
    csv_file = csv.reader(f)
    data = list(csv_file)
    # Assuming first row is header
    
    data = data[1:]  # Skip header

    specific = data[0][:-1]
    general = [['?' for _ in range(len(specific))] for _ in range(len(specific))]

    for i in data:
        if i[-1] == "Yes":
            for j in range(len(specific)):
                if i[j] != specific[j]:
                    specific[j] = "?"
                    general[j][j] = "?"
        elif i[-1] == "No":
            for j in range(len(specific)):
                if i[j] != specific[j]:
                    general[j][j] = specific[j]
                else:
                    general[j][j] = "?"

    # Filter general hypotheses to include only those that aren't all '?'
    gh = [g for g in general if any(val != '?' for val in g)]

    # Final Output
    print("Final Specific Hypothesis:\n", specific)
    print("\nFinal General Hypotheses:")
    for g in gh:
        print(g)
```

```python
#ex-3
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/bhatt941/Tennis-Dataset/refs/heads/main/PlayTennis.csv"
df = pd.read_csv(url)


print(df.head())

df['Outlook'] = df['Outlook'].map({'Sunny': 0, 'Overcast': 1, 'Rain': 2})
df['Temperature'] = df['Temperature'].map({'Hot': 0, 'Mild': 1, 'Cool': 2})
df['Humidity'] = df['Humidity'].map({'High': 0, 'Normal': 1})
df['Wind'] = df['Wind'].map({'Weak': 0, 'Strong': 1})
df['Play Tennis'] = df['Play Tennis'].map({'No': 0, 'Yes': 1})

X = df[['Outlook', 'Temperature', 'Humidity', 'Wind']]
y = df['Play Tennis']

clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)

plt.figure(figsize=(20, 10))
plot_tree(clf, feature_names=['Outlook', 'Temperature', 'Humidity', 'Wind'],class_names=['No', 'Yes'], filled=True)
plt.show()

new_sample = [[0, 1, 0, 1]]  
prediction = clf.predict(new_sample)
print("Predicted class:", 'Yes' if prediction[0] == 1 else 'No')
```

```python
#ex-7
import numpy as np

# AND Gate Input and Output
X = np.array([[0,0], [0,1], [1,0], [1,1]])  # inputs
y = np.array([[0], [0], [0], [1]])          # expected output

# Initialize weights and bias
weights = np.random.rand(2, 1)
bias = np.random.rand(1)

# Sigmoid activation function and derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x):
    return x * (1 - x)

# Training the network
for epoch in range(10000):
    z = np.dot(X, weights) + bias      # Linear combination
    pred = sigmoid(z)                  # Activation

    error = y - pred                   # Difference between actual and predicted
    adjustments = error * sigmoid_derivative(pred)  # Gradient
    
    # Update weights and bias
    weights += np.dot(X.T, adjustments)
    bias += np.sum(adjustments)

# Testing the model
print("Predictions after training:")
for input_data in X:
    result = sigmoid(np.dot(input_data, weights) + bias)
    print(f"Input: {input_data}, Predicted: {round(float(result), 2)}")
```

```python
#ex-8
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create k-NN model (k=3)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Make predictions
y_pred = knn.predict(X_test)

# Compare predictions with actual labels
print("Correct and Incorrect Predictions:\n")
for i in range(len(y_test)):
    actual = iris.target_names[y_test.values[i]]
    predicted = iris.target_names[y_pred[i]]
    status = "Correct " if y_pred[i] == y_test.values[i] else "Wrong "
    print(f"Sample {i+1}: Actual = {actual:10s} | Predicted = {predicted:10s} --> {status}")

# Print final accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nOverall Accuracy: {accuracy*100:.2f}%")
```

```python
#ex-9
import numpy as np
import matplotlib.pyplot as plt
# Generate sample dataset
np.random.seed(42)
X = np.linspace(0, 10, 100)
y = np.sin(X) + np.random.normal(scale=0.1, size=X.shape)

# Convert to column vector
X = X.reshape(-1, 1)
def weights(X, x_query, tau):
    return np.exp(-np.sum((X - x_query) ** 2, axis=1) / (2 * tau ** 2))
def locally_weighted_regression(X, y, x_query, tau=0.5):
    W = np.diag(weights(X, x_query, tau))
    X_bias = np.hstack((np.ones_like(X), X))
    theta = np.linalg.pinv(X_bias.T @ W @ X_bias) @ X_bias.T @ W @ y
    return np.array([1, x_query]) @ theta
# Fit model
X_test = np.linspace(0, 10, 100)
y_pred = np.array([locally_weighted_regression(X, y, x, tau=0.5) for x in X_test])
# Plot results
plt.scatter(X, y, label='Data', alpha=0.6)
plt.plot(X_test, y_pred, color='red', label='LWR Fit')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Locally Weighted Regression')
plt.show()
```

```python
#ex-10
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score
# Expanded sample text dataset
documents = [
    "This is a sports article", "Politics is complicated", "I love watching football", 
    "Government policies affect economy", "Sports events are exciting", "The new election results are out", 
    "The team won the championship", "New policies affect trade", "Soccer is a popular sport", 
    "The parliament passed a new bill"
]
labels = ["Sports", "Politics", "Sports", "Politics", "Sports", "Politics", "Sports", "Politics", "Sports", "Politics"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)
y = labels
# Split dataset with stratification to ensure label balance
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
# Train Nave Bayes Classifier
nb = MultinomialNB()
nb.fit(X_train, y_train)
# Predictions
y_pred = nb.predict(X_test)
# Evaluate Model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)
recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
```

```python
#ex-11
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Load dataset
url = 'https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/refs/heads/master/dataset.csv'
df = pd.read_csv(url)

# Step 2: Preprocessing
features = df.drop(columns=['target'])  # remove target for unsupervised learning
scaler = StandardScaler()
X = scaler.fit_transform(features)

# Step 3: K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
kmeans_score = silhouette_score(X, kmeans_labels)

# Step 4: EM Algorithm (Gaussian Mixture)
gmm = GaussianMixture(n_components=2, random_state=42)
gmm_labels = gmm.fit_predict(X)
gmm_score = silhouette_score(X, gmm_labels)

# Step 5: Output results
print("=== Clustering Evaluation ===")
print(f"K-Means Silhouette Score: {kmeans_score:.4f}")
print(f"EM (GMM) Silhouette Score: {gmm_score:.4f}")

# Optional: compare clusters with target labels
df['KMeans_Cluster'] = kmeans_labels
df['GMM_Cluster'] = gmm_labels
print("\nSample cluster assignments:")
print(df[['target', 'KMeans_Cluster', 'GMM_Cluster']].head())
```

```python
#ex-14
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
# Example dataset
data_dict = {
    'Feature1': [2.5, 0.5, 2.2, 1.9, 3.1],
    'Feature2': [2.4, 0.7, 2.9, 2.2, 3.0],
    'Feature3': [2.7, 0.8, 2.5, 2.3, 3.1],
    'Label': [0, 1, 0, 1, 0]  # Labels for classification (0 or 1)
}
data = pd.DataFrame(data_dict)
print("Original Data:")
print(data)
# Separate features and labels
X = data.drop('Label', axis=1)
y = data['Label']
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Perform PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
X_pca = pca.fit_transform(X_scaled)
print("\nPrincipal Components:")
print(X_pca)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
# Train the SVM model
svm = SVC(kernel='linear')  # Linear kernel
svm.fit(X_train, y_train)
# Predict and evaluate the model
y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("\nAccuracy of the SVM model:", accuracy)
```

```python
#ex-13
import pandas as pd
import numpy as np
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import BayesianEstimator
from pgmpy.inference import VariableElimination
# Load dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"
columns = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]
df = pd.read_csv(url, names=columns, na_values="?")
df.dropna(inplace=True)
# Discretize numerical columns to reduce memory usage
df['age'] = pd.cut(df['age'], bins=[0, 40, 55, 65, 100], labels=[0, 1, 2, 3])
df['chol'] = pd.cut(df['chol'], bins=[0, 200, 240, 600], labels=[0, 1, 2])
df['thalach'] = pd.cut(df['thalach'], bins=[0, 120, 150, 210], labels=[0, 1, 2])
df = df.astype(int)  # Convert categories to numerical values
# Define Bayesian Network Structure
model = DiscreteBayesianNetwork([
    ("cp", "target"),
    ("thal", "target"),
    ("ca", "target"),
    ("exang", "target"),
    ("slope", "target"),
    ("chol", "target"),
    ("thalach", "target")
])
# Learn CPDs with Bayesian Estimator
model.fit(df, estimator=BayesianEstimator, prior_type="BDeu")
# Perform inference
inference = VariableElimination(model)
# Example: Diagnosing a patient with given symptoms
evidence = {"cp": 2, "chol": 1, "thalach": 1}
query_result = inference.map_query(variables=["target"], evidence=evidence)
print("Predicted Heart Disease Presence (0 = No, 1 = Yes):", query_result["target"])
```
